```python
# csv
dfbk = pd.read_csv(r'data\brooklyn_tosubplace.csv')
dfbk = dfbk.sort_values(by ='Date')
dfbk['Borough'] = 'Brooklyn'

dfbx = pd.read_csv(r'data\bronx_tosubplace.csv')
dfbx = dfbx.sort_values(by ='Date')
dfbx['Borough'] = 'Bronx'

dfmn = pd.read_csv(r'data\manhattan_tosubplace.csv')
dfmn = dfmn.sort_values(by ='Date')
# dfmn = dfmn.drop(dfmn.columns[0], axis=1)
dfmn['Borough'] = 'Manhattan'

dfqn = pd.read_csv(r'data\queens_tosubplace.csv')
dfqn = dfqn.sort_values(by ='Date')
dfqn['Borough'] = 'Queens'

dfsi = pd.read_csv(r'data\staten_tosubplace.csv')
dfsi = dfsi.sort_values(by ='Date')
dfsi['Borough'] = 'Staten Island'

df = pd.concat([dfbk, dfbx, dfmn, dfqn, dfsi], ignore_index =True)
#df = df.sort_values(by ='Borough')
# geojson
#subplaces = json.load(open(r'data\shapefiles\nyc_subplace_centroid.geojson','r'))
```

```python
us19 = pd.read_csv(r'data\mn_to_usa19_adj.csv')
us19 = us19.sort_values(by ='date_viz')
us19['Year'] = '2019'

us20 = pd.read_csv(r'data\mn_to_usa20_adj.csv')
us20 = us20.sort_values(by ='date_viz')
us20['Year'] = '2020'
df_us = pd.concat([us19,us20], ignore_index =True)
df_us = df_us.sort_values(by = 'date_viz')
```

```python
fig = px.scatter_mapbox(df,# geojson = subplaces,
                     lat = 'INTPTLAT', lon = 'INTPTLON',
                     size = 'Trips/Device',
                     color ='Borough',
                     color_discrete_sequence = px.colors.qualitative.Safe,
                     animation_frame = 'Date',
                     animation_group = 'Borough',
                     hover_data = {'Name':True,'Date':True,'Trips/Device':True, 'INTPTLAT':False, 'INTPTLON':False, 'Borough':False},
                #     labels={'Borough':'Borough of Residence'},
                     #font = dict(family = 'Arial', color = 'white'),
                     size_max = 60)
           #          labels = 'Borough')
fig.update_layout(mapbox_zoom = 6.8,
                 mapbox_center = {'lat':40.82, 'lon':-74.25},
                 mapbox_style="carto-darkmatter",
                 width = 950,
                 height = 780, 
                 font = dict(family = 'Arial', color = 'black'),
                 title_font_color = 'white',
                 legend=dict(
#                       orientation="h",
                      title = 'Borough of Residence',
                      title_font_color = 'white',
                      bgcolor = 'black',
                      bordercolor = 'gray', borderwidth = 1,
                      font = dict(family = 'Arial', color = 'white'),
                      yanchor="top",
                      y=0.33,
                      xanchor="left",
                      x=0.02))#,
   #              margin = {'r':1,'t':1,'l':1,'b':1},
  #               legend = dict(yanchor = "top", y = 0.8, xanchor = "left", x = 0.10))
fig.add_annotation(text = "Where NYC Residents Traveled in the<br>Surrounding Metropolitan Region",
                  align = 'left', x = 0.01, y = 0.95, showarrow = False,
                  bordercolor = None,
                  bgcolor = None,
                  font = dict(family = "Arial", color = 'white', size = 19))
fig.add_annotation(text = 'January 1, 2020 to September 30, 2020',
                  align = 'left', x = 0.01, y = 0.86, showarrow = False,
                  bordercolor = None,
                  bgcolor = None,
                  font = dict(family = "Arial", color = 'white', size = 13))
fig.add_annotation(text = 'Bubble size = Trips per 100,000 mobile devices',
                  align = 'right', x = 0.02, y = 0.082, showarrow=False,
                  bordercolor = None, bgcolor = None,
                  font = dict(family = 'Droid Sans', color = 'white', size = 12))
fig.add_annotation(text = 'Click on/off boroughs in legend to compare or isolate migratory patterns',
                  align = 'right', x = 0.02, y = 0.01, showarrow=False,
                  bordercolor = None, bgcolor = None,
                  font = dict(family = 'Droid Sans', color = 'white', size = 12))
                  
#fig.show()
fig.write_html('subplace_nyc.html', auto_open = True, auto_play = False, include_plotlyjs = 'cdn')
#title top left corner
# under legend, comment for text(annotation)
```
